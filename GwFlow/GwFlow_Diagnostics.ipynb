```python
import numpy as np
import matplotlib.pyplot as plt

from diagnostics import *
```

```python
pars_with = np.loadtxt('parameters_with.csv')
pars_without = np.loadtxt('parameters_without.csv')
```

```python
n_links = 20000
n_chains = 2

parameters_with = []
parameters_without = []
for i in range(n_chains):
    parameters_with.append(pars_with[i*n_links:(i+1)*n_links,:])
    parameters_without.append(pars_without[i*n_links:(i+1)*n_links,:])
```

```python
for i in range(n_chains):
    plot_parameters(parameters_with[i], [0,1,2,3,4])
```

```python
compute_R_hat(parameters_with)
```

```python
compute_ESS(parameters_with).mean()
```

```python
T = 300
a = get_autocorrelation(parameters_with[0][:,0], T=T)
x = np.arange(0, a.shape[0])
plt.figure(figsize=(12,8))
plt.plot(x, a, color='k', linewidth=2)
plt.fill_between(x, a, color='k', alpha=0.2)
plt.xlabel('Lag', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('acf_with_aem.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
for i in range(n_chains):
    plot_parameters(parameters_without[i], [0,1,2,3,4])
```

```python
compute_R_hat(parameters_without)
```

```python
compute_ESS(parameters_without).mean()
```

```python
T = 300
a = get_autocorrelation(parameters_without[0][:,0], T=T)
x = np.arange(0, a.shape[0])
plt.figure(figsize=(12,8))
plt.plot(x, a, color='k', linewidth=2)
plt.fill_between(x, a, color='k', alpha=0.2)
plt.xlabel('Lag', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('acf_without_aem.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
(compute_ESS(parameters_with)/compute_ESS(parameters_without)).mean()
```
